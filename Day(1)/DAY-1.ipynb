```python
# Importing Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer

# Load Dataset
df = pd.read_csv("C:/Users/kondu/Downloads/day1/mental_health_workplace_survey.csv")

#Data Cleaning

# Checking for missing values
print("Missing Values in Dataset:")
print(df.isnull().sum())

# Fill missing values (Numerical columns with mean, Categorical columns with mode)
imputer = SimpleImputer(strategy='mean')

# List of numerical columns for imputation
numerical_cols = ['Age', 'WorkHoursPerWeek', 'StressLevel', 'ProductivityScore', 'SleepHours', 'PhysicalActivityHrs', 'ManagerSupportScore', 'MentalHealthDaysOff', 'WorkLifeBalanceScore', 'CareerGrowthScore']

# Apply imputation
df[numerical_cols] = imputer.fit_transform(df[numerical_cols])

# For categorical columns (e.g., 'RemoteWork', 'HasMentalHealthSupport')
categorical_cols = ['RemoteWork', 'HasMentalHealthSupport', 'HasTherapyAccess', 'SalaryRange']
for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

#Handling Outliers
df = df[df['WorkHoursPerWeek'] <= 80]

#Exploratory Data Analysis (EDA)

# Visualizing relationships
sns.scatterplot(x='StressLevel', y='SleepHours', data=df)
plt.title('Stress Level vs Sleep Hours')
plt.show()

sns.boxplot(x='BurnoutLevel', y='WorkHoursPerWeek', data=df)
plt.title('Burnout Level vs Work Hours per Week')
plt.show()

# Display cleaned data and checki if there are any remaining missing values
print("Cleaned Data:")
print(df.isnull().sum())
```

```python
# Save the cleaned dataset
df.to_csv("C:/Users/kondu/Downloads/day1/day1_updated_mental_health.csv", index=False)

print(" Cleaned dataset saved as 'day1_updated_mental_health.csv'")
```
